```python
# Cell 2: Embed enos.html
import base64, os
os.makedirs("books", exist_ok=True)
html_b64 = ""
with open("books/enos.html", "wb") as f:
    f.write(base64.b64decode(html_b64))
print(f"Written {os.path.getsize("books/enos.html")} bytes")
```

```python
# Cell 3: Generate audio
import json, os, re, time
import numpy as np
from bs4 import BeautifulSoup, NavigableString
from pathlib import Path

VOICE = "bm_george"
SAMPLE_RATE = 24000
LINE_PAUSE_S = 0.18
VERSE_PAUSE_S = 0.50
PERICOPE_PAUSE_S = 0.90

def get_text_from_element(el, use_modern=True):
    clone = BeautifulSoup(str(el), "html.parser")
    if use_modern:
        for swap in clone.find_all(class_=["swap", "swap-quiet"]):
            mod = swap.get("data-mod")
            if mod:
                swap.string = mod
    for remove in clone.find_all(class_=["verse-num", "parry-label-spacer", "parry-label"]):
        remove.decompose()
    text = clone.get_text()
    text = re.sub(r"\s+", " ", text).strip()
    return text

def parse_chapter(html_path, book_id, chapter_num):
    with open(html_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f.read(), "html.parser")
    chapter_id = f"ch-{book_id}-{chapter_num}"
    chapter_div = soup.find(id=chapter_id)
    if not chapter_div:
        print(f"Warning: chapter {chapter_id} not found")
        return []
    items = []
    line_idx = 0
    for child in chapter_div.children:
        if isinstance(child, NavigableString):
            continue
        classes = child.get("class", [])
        if "pericope-header" in classes:
            main_el = child.find(class_="pericope-main")
            if main_el:
                text = main_el.get_text().strip()
                if text:
                    items.append({"type": "pericope", "text": text, "verse_num": "", "line_index": line_idx})
                    line_idx += 1
            continue
        if "verse" in classes:
            verse_num_el = child.find(class_="verse-num")
            vn = verse_num_el.get_text().strip() if verse_num_el else ""
            sense_lines = [el for el in child.find_all("span", recursive=False) if el.get("class") == ["line"]]
            for line_el in sense_lines:
                text = get_text_from_element(line_el, use_modern=True)
                if text:
                    items.append({"type": "line", "text": text, "verse_num": vn, "line_index": line_idx})
                    line_idx += 1
            if items and items[-1]["type"] != "verse-gap":
                items.append({"type": "verse-gap", "text": "", "verse_num": "", "line_index": -1})
    return items

# Parse
items = parse_chapter("books/enos.html", "enos", 1)
speakable = [i for i in items if i["type"] in ("pericope", "line")]
print(f"Found {len(speakable)} speakable items")
for item in speakable[:5]:
    print(f"  [{item["type"]}] {item["text"][:60]}...")
```

```python
# Cell 4: Load Kokoro and generate audio
from kokoro import KPipeline
print("Loading Kokoro TTS model...")
tts = KPipeline(lang_code="a")
print("Model loaded!")

all_samples = []
timing = []
current_time = 0.0
total = len(speakable)
done = 0

def gen_silence(dur):
    return np.zeros(int(dur * SAMPLE_RATE), dtype=np.float32)

for item in items:
    if item["type"] == "verse-gap":
        s = gen_silence(VERSE_PAUSE_S)
        all_samples.append(s)
        current_time += VERSE_PAUSE_S
        continue
    if item["type"] == "pericope":
        s = gen_silence(PERICOPE_PAUSE_S)
        all_samples.append(s)
        current_time += PERICOPE_PAUSE_S
    start_time = current_time
    try:
        result = tts.generate(item["text"], voice=VOICE)
        audio_data = result.audio
        if hasattr(audio_data, "numpy"):
            audio_data = audio_data.numpy()
        audio_data = np.array(audio_data, dtype=np.float32).flatten()
    except Exception as e:
        print(f"Error: {e}")
        audio_data = gen_silence(0.5)
    all_samples.append(audio_data)
    duration = len(audio_data) / SAMPLE_RATE
    current_time += duration
    timing.append({"start": round(start_time, 3), "end": round(current_time, 3), "type": item["type"], "text": item["text"], "verse": item["verse_num"], "lineIndex": item["line_index"]})
    done += 1
    print(f"[{done}/{total}] {round(done/total*100)}%", end="", flush=True)
    if item["type"] == "line":
        all_samples.append(gen_silence(LINE_PAUSE_S))
        current_time += LINE_PAUSE_S
    if item["type"] == "pericope":
        all_samples.append(gen_silence(PERICOPE_PAUSE_S))
        current_time += PERICOPE_PAUSE_S

combined = np.concatenate(all_samples)
print(f"
Done! Total audio: {len(combined)/SAMPLE_RATE:.1f}s")
```

```python
# Cell 5: Encode to MP3, save manifest, and download
import lameenc
from google.colab import files

# MP3 encode
pcm_16 = (combined * 32767).astype(np.int16)
encoder = lameenc.Encoder()
encoder.set_bit_rate(64)
encoder.set_in_sample_rate(SAMPLE_RATE)
encoder.set_channels(1)
encoder.set_quality(2)
mp3_data = encoder.encode(pcm_16.tobytes())
mp3_data += encoder.flush()

with open("enos-1.mp3", "wb") as f:
    f.write(mp3_data)
print(f"MP3: {len(mp3_data)/1024:.0f} KB")

# Timing manifest
manifest = {"book": "enos", "bookName": "Enos", "chapter": 1, "voice": VOICE, "duration": round(len(combined)/SAMPLE_RATE, 3), "lines": timing}
with open("enos-1.json", "w") as f:
    json.dump(manifest, f, indent=2)
print(f"Manifest: {len(timing)} entries")

# Download both files
files.download("enos-1.mp3")
files.download("enos-1.json")
print("Downloads triggered!")
```
